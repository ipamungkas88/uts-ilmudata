Visualisasi

```python
# Create a performance score based on CPU and GPU specifications
# Performance score = (cpu_tier + gpu_tier + cpu_cores/4 + vram_gb/2 + ram_gb/8) * cpu_boost_ghz
data['performance_score'] = (data['cpu_tier'] + data['gpu_tier'] + 
                           data['cpu_cores']/4 + data['vram_gb']/2 + 
                           data['ram_gb']/8) * data['cpu_boost_ghz']

print("Performance score created successfully!")
print(f"Performance score range: {data['performance_score'].min():.2f} - {data['performance_score'].max():.2f}")
print(f"Average performance score: {data['performance_score'].mean():.2f}")
data[['brand', 'model', 'price', 'performance_score']].head()
```

```python
# 1. Scatter Plot: Harga vs Performance Score
plt.figure(figsize=(12, 8))
plt.scatter(data['performance_score'], data['price'], alpha=0.6, s=50)
plt.xlabel('Performance Score', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.title('Scatter Plot: Price vs Performance Score', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(data['performance_score'], data['price'], 1)
p = np.poly1d(z)
plt.plot(data['performance_score'], p(data['performance_score']), "r--", alpha=0.8, linewidth=2)

# Add correlation coefficient
correlation = data['performance_score'].corr(data['price'])
plt.text(0.05, 0.95, f'Correlation: {correlation:.3f}', transform=plt.gca().transAxes, 
         bbox=dict(boxstyle="round", facecolor='wheat', alpha=0.8), fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# 2. Bar Chart: Top 10 Brand dengan Harga Rata-rata Tertinggi
plt.figure(figsize=(12, 8))

# Calculate average price by brand
avg_price_by_brand = data.groupby('brand')['price'].mean().sort_values(ascending=False).head(10)

# Create bar chart
bars = plt.bar(range(len(avg_price_by_brand)), avg_price_by_brand.values, 
               color=sns.color_palette("viridis", len(avg_price_by_brand)))

# Customize the chart
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.title('Top 10 Brands with Highest Average Price', fontsize=14, fontweight='bold')
plt.xticks(range(len(avg_price_by_brand)), avg_price_by_brand.index, rotation=45, ha='right')

# Add value labels on top of bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 20,
             f'${height:.0f}', ha='center', va='bottom', fontweight='bold')

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Print summary statistics
print("\nTop 10 Brands by Average Price:")
for i, (brand, price) in enumerate(avg_price_by_brand.items(), 1):
    print(f"{i:2d}. {brand:15s}: ${price:6.2f}")
```

```python
# 3. Box Plot: Sebaran Harga Berdasarkan OS
plt.figure(figsize=(12, 8))

# Create box plot
sns.boxplot(data=data, x='os', y='price', palette='Set2')

# Customize the chart
plt.xlabel('Operating System', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.title('Price Distribution by Operating System', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)

# Add statistical annotations
os_stats = data.groupby('os')['price'].agg(['count', 'mean', 'median', 'std']).round(2)
print("Price Statistics by Operating System:")
print(os_stats)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 4. Histogram: Distribusi Harga
plt.figure(figsize=(12, 8))

# Create histogram with density curve
plt.hist(data['price'], bins=50, alpha=0.7, color='skyblue', edgecolor='black', density=True)

# Add density curve
from scipy import stats
density = stats.gaussian_kde(data['price'])
xs = np.linspace(data['price'].min(), data['price'].max(), 200)
plt.plot(xs, density(xs), color='red', linewidth=2, label='Density Curve')

plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Price Distribution Histogram', fontsize=14, fontweight='bold')
plt.legend()

# Add statistical annotations
mean_price = data['price'].mean()
median_price = data['price'].median()
plt.axvline(mean_price, color='red', linestyle='--', alpha=0.8, label=f'Mean: ${mean_price:.0f}')
plt.axvline(median_price, color='green', linestyle='--', alpha=0.8, label=f'Median: ${median_price:.0f}')

plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Price Statistics:")
print(f"Mean: ${data['price'].mean():.2f}")
print(f"Median: ${data['price'].median():.2f}")
print(f"Standard Deviation: ${data['price'].std():.2f}")
print(f"Min: ${data['price'].min():.2f}")
print(f"Max: ${data['price'].max():.2f}")
```

```python
# 5. Pie Chart: Distribusi Device Type
plt.figure(figsize=(10, 8))

# Calculate device type distribution
device_counts = data['device_type'].value_counts()
print(f"Number of device types: {len(device_counts)}")
print("Device types:", list(device_counts.index))

# Create pie chart with appropriate number of colors and explode values
colors = sns.color_palette("husl", len(device_counts))
explode_values = [0.05 if i == 0 else 0 for i in range(len(device_counts))]

wedges, texts, autotexts = plt.pie(device_counts.values, labels=device_counts.index, 
                                   autopct='%1.1f%%', colors=colors, startangle=90,
                                   explode=explode_values)

# Customize the chart
plt.title('Distribution of Device Types', fontsize=14, fontweight='bold')

# Enhance text appearance
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)

for text in texts:
    text.set_fontsize(12)
    text.set_fontweight('bold')

# Add total count information
total = device_counts.sum()
plt.text(0, -1.3, f'Total Devices: {total:,}', ha='center', fontsize=12, fontweight='bold')

plt.axis('equal')
plt.tight_layout()
plt.show()

# Print detailed statistics
print("\nDevice Type Distribution:")
for device_type, count in device_counts.items():
    percentage = (count / total) * 100
    print(f"{device_type}: {count:,} ({percentage:.1f}%)")
```

```python
# 6. Heatmap: Correlation Matrix untuk variabel numerik
plt.figure(figsize=(14, 10))

# Select numeric columns for correlation
numeric_cols = ['price', 'performance_score', 'cpu_cores', 'cpu_threads', 'ram_gb', 
               'storage_gb', 'vram_gb', 'cpu_base_ghz', 'cpu_boost_ghz', 'release_year']

# Calculate correlation matrix
corr_matrix = data[numeric_cols].corr()

# Create heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Mask upper triangle
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='RdYlBu_r', center=0,
            square=True, fmt='.3f', cbar_kws={"shrink": .8})

plt.title('Correlation Matrix of Numeric Variables', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Print strongest correlations with price
price_correlations = corr_matrix['price'].abs().sort_values(ascending=False)
print("Strongest correlations with price:")
for var, corr in price_correlations.items():
    if var != 'price':
        print(f"{var:20s}: {corr:.3f}")
```

```python
# 7. Bar Chart: Harga Rata-rata berdasarkan CPU Brand
plt.figure(figsize=(12, 8))

# Calculate average price by CPU brand
avg_price_by_cpu = data.groupby('cpu_brand')['price'].mean().sort_values(ascending=False)

# Create bar chart
bars = plt.bar(range(len(avg_price_by_cpu)), avg_price_by_cpu.values, 
               color=sns.color_palette("Set1", len(avg_price_by_cpu)))

# Customize the chart
plt.xlabel('CPU Brand', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.title('Average Price by CPU Brand', fontsize=14, fontweight='bold')
plt.xticks(range(len(avg_price_by_cpu)), avg_price_by_cpu.index, rotation=45)

# Add value labels on top of bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 20,
             f'${height:.0f}', ha='center', va='bottom', fontweight='bold')

# Add statistical summary
cpu_brand_stats = data.groupby('cpu_brand')['price'].agg(['count', 'mean', 'median', 'std']).round(2)
print("Price Statistics by CPU Brand:")
print(cpu_brand_stats)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 8. Line Chart: Trend Harga Rata-rata berdasarkan Tahun Rilis
plt.figure(figsize=(12, 8))

# Calculate average price by release year
avg_price_by_year = data.groupby('release_year').agg({
    'price': ['mean', 'count', 'std']
}).round(2)

avg_price_by_year.columns = ['avg_price', 'count', 'std_price']
avg_price_by_year = avg_price_by_year.reset_index()

# Create simple line chart
plt.plot(avg_price_by_year['release_year'], avg_price_by_year['avg_price'], 
         marker='o', markersize=6, linewidth=2, color='blue')

# Customize the chart
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.title('Average Price Trend by Release Year', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print year-by-year statistics
print("Average Price by Release Year:")
for _, row in avg_price_by_year.iterrows():
    print(f"{int(row['release_year'])}: ${row['avg_price']:6.2f} (±${row['std_price']:5.2f}) - {int(row['count'])} devices")
```

```python
# 9. Stacked Bar Chart: Komposisi GPU Brand berdasarkan Device Type
plt.figure(figsize=(12, 8))

# Create cross-tabulation
gpu_device_crosstab = pd.crosstab(data['device_type'], data['gpu_brand'], normalize='index') * 100

# Create stacked bar chart
ax = gpu_device_crosstab.plot(kind='bar', stacked=True, figsize=(12, 8), 
                             color=sns.color_palette("Set2", len(gpu_device_crosstab.columns)))

# Customize the chart
plt.xlabel('Device Type', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.title('GPU Brand Composition by Device Type', fontsize=14, fontweight='bold')
plt.legend(title='GPU Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)

# Add percentage labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center', fontweight='bold')

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Print detailed statistics
print("GPU Brand Distribution by Device Type (%):")
print(gpu_device_crosstab.round(1))
print("\nActual Counts:")
print(pd.crosstab(data['device_type'], data['gpu_brand']))
```
